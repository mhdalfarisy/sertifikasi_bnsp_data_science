• **Potensi Keuntungan**

Dalam bisnis properti, penentuan harga rumah yang tidak tepat dapat menyebabkan dua kondisi utama, yaitu **underprice** dan **overprice**.

- **Underprice (harga terlalu rendah)** dapat menyebabkan rumah dijual di bawah nilai pasar sehingga mengurangi potensi keuntungan penjualan.
- **Overprice (harga terlalu tinggi)** dapat menyebabkan rumah kurang kompetitif di pasar dan membutuhkan waktu lebih lama untuk terjual.

Dengan memanfaatkan model machine learning untuk memprediksi harga rumah berdasarkan faktor geografis, demografis, dan karakteristik properti, perusahaan dapat memperoleh rekomendasi harga yang lebih mendekati nilai pasar.

Strategi penyesuaian harga yang dapat diterapkan antara lain:

- Jika harga rumah terindikasi **lebih rendah dari nilai pasar (underprice)**, maka harga dapat dinaikkan sekitar **5%** untuk mengoptimalkan keuntungan.
- Jika harga rumah terindikasi **lebih tinggi dari nilai pasar (overprice)**, maka harga dapat diturunkan sekitar **2–3%** agar lebih kompetitif di pasar.

Sebagai ilustrasi, jika rata-rata harga rumah berada pada kisaran **$200,000**, maka peningkatan harga sebesar **5%** dapat memberikan tambahan nilai sekitar **$10,000 per rumah**.

Jika strategi ini diterapkan pada **50 rumah dalam satu tahun**, maka potensi peningkatan nilai transaksi dapat mencapai sekitar:

**$10,000 × 50 = $500,000**

```python
# Plot boxplot untuk mengecek outliers
df.plot(kind='box', figsize=(20,15), subplots=True, layout=(5,4))
plt.show()
```

```python
def check_outliers(df,col):
    Q1=df[col].describe()['25%']
    Q3=df[col].describe()['75%']
    IQR=Q3-Q1
    Upper_fence=Q3+(1.5*IQR)
    Lower_fence=Q1-(1.5*IQR)
    df_outliers=df[(df[col]<Lower_fence) | (df[col]>Upper_fence)]
    if len(df_outliers)==0:
        return 'Tidak Ada Outliers'
    else:
        return df_outliers
```

```python
sns.boxplot(df['housing_median_age'])
plt.title('Boxplot Outlier housing_median_age')
plt.show()
outliers_housing_median_age=check_outliers(df,'housing_median_age')
outliers_housing_median_age
```

```python
# Encoding Kategori
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
```

```python
# Kolom kategorikal dan numerikal
categorical_column = ['ocean_proximity']
numerical_column = [
    'housing_median_age',
    'total_rooms',
    'total_bedrooms',
    'population',
    'households',
    'median_income'
]

# Pipeline untuk kolom kategorikal
categorical_tranformer = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),  # Isi NaN dengan modus
    ('onehot', OneHotEncoder(drop='first', handle_unknown='ignore'))
])

# Pipeline untuk kolom numerikal
numerical_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler(with_mean=False))
])

# Gabungkan pipeline dengan ColumnTransformer
transformer = ColumnTransformer([
    ('cat', categorical_tranformer, categorical_column),
    ('num', numerical_transformer, numerical_column)
], remainder='passthrough')
```

##### Pipeline Kategorikal dan Numerikal

##### 1. Kolom Kategorikal dan Numerikal
- **Kolom Kategorikal**: `ocean_proximity`  
  → Berisi informasi kategori tentang lokasi rumah relatif terhadap laut.
- **Kolom Numerikal**: `housing_median_age`, `total_rooms`, `total_bedrooms`, `population`, `households`, `median_income`  
  → Berisi informasi numerik seperti jumlah kamar, usia rumah, jumlah penduduk, dan pendapatan.

##### 2. Pipeline Kolom Kategorikal
- **Imputasi**:  
  Mengisi nilai yang hilang (NaN) dengan modus (kategori yang paling sering muncul).  
  Hal ini dilakukan karena **OneHotEncoder tidak bisa memproses nilai kosong**.
- **Encoding**:  
  Mengubah kategori menjadi angka menggunakan **OneHotEncoding**.  
  - `drop='first'` digunakan untuk menghindari dummy variable trap, yaitu kolinearitas antara kolom dummy.  
  - `handle_unknown='ignore'` digunakan agar pipeline tidak error jika muncul kategori baru saat prediksi.

##### 3. Pipeline Kolom Numerikal
- **Imputasi**:  
  Mengisi nilai hilang (NaN) dengan **median**.  
  Median dipilih karena distribusi kolom numerikal biasanya **tidak normal atau skewed**, sehingga median lebih representatif daripada mean.
- **Scaling**:  
  Menormalkan nilai numerik menggunakan **StandardScaler** agar skala setiap kolom setara.  
  Hal ini membantu model machine learning untuk **tidak bias terhadap fitur dengan skala besar**.  
  - `with_mean=False` dipakai agar pipeline aman jika nanti dikombinasikan dengan sparse matrix hasil encoding.


##### 4. Menggabungkan Pipeline
- Pipeline kategorikal dan numerikal digabungkan menggunakan **ColumnTransformer**.  
- Setiap pipeline diterapkan hanya pada kolom yang sesuai, sementara kolom lain tetap dilewatkan (`remainder='passthrough'`).  
- Hasilnya adalah **data siap pakai untuk model machine learning**, di mana semua fitur numerik telah diskalakan dan kategori telah diubah menjadi angka.

### Ringkasan
1. Data kategori → diisi NaN dengan modus → diubah menjadi dummy variables  
2. Data numerik → diisi NaN dengan median → diskalakan agar setara  
3. Semua pipeline digabung → memudahkan preprocessing sebelum training model

```python
!pip install xgboost

# Model Machine Learning
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor

# Target Transformation
from sklearn.compose import TransformedTargetRegressor

# Evaluation Metrics
from sklearn.metrics import (
    mean_absolute_error,
    mean_absolute_percentage_error
)

# Model Selection & Validation
from sklearn.model_selection import (
    train_test_split,
    cross_val_score,
    RandomizedSearchCV,
    GridSearchCV,
    KFold,
    cross_validate
)
```

```python
## Memisahkan data independen variabel dengan target
X = df.drop(['median_house_value'], axis=1)
y = df['median_house_value']
```

```python
## Splitting data training dan test dengan proporsi 70:30
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
## Menyiapkan algoritma yang digunakan
model_lr = LinearRegression()
model_knn = KNeighborsRegressor()
model_dt = DecisionTreeRegressor(random_state=42)
model_rf = RandomForestRegressor(random_state=42)
model_xgb = XGBRegressor(random_state=42)
```

```python
## Pemodelan dilakukan dalam skala logaritmik, namun kemudian di-inverse kembali untuk interpretasi
log_lr = TransformedTargetRegressor(model_lr, func=np.log, inverse_func=np.exp)
log_knn = TransformedTargetRegressor(model_knn, func=np.log, inverse_func=np.exp)
log_dt = TransformedTargetRegressor(model_dt, func=np.log, inverse_func=np.exp)
log_rf = TransformedTargetRegressor(model_rf, func=np.log, inverse_func=np.exp)
log_xgb = TransformedTargetRegressor(model_xgb, func=np.log, inverse_func=np.exp)
```

```python
## Memasukan kandidat algoritma dalam sebuah variabel
models = [log_lr, log_knn, log_dt, log_rf, log_xgb]

## Untuk menyimpan nilai MAE
score_mae = []
nilai_mean_mae = []
nilai_std_mae = []

## Untuk menyimpan nilai MAPE
score_mape = []
nilai_mean_mape = []
nilai_std_mape = []


## Mencari algoritma terbaik dalam tahap train
for i in models:
    
    ## metode statistik yang dapat digunakan untuk mengevaluasi kinerja model
    crossval = KFold(n_splits=5, shuffle=True, random_state=42) #n_split 5 kombinasi, #shuffle untuk mengacak -> dibagi 5 berdasarkan n_splits


    estimator = Pipeline([
        ('preprocessing', transformer),
        ('model_machine_learning', i)
    ])

    ## MAE
    model_cv_mae = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_mean_absolute_error',  
        error_score='raise' ### Jika diatur ke 'raise', nilai error dimunculkan 
        )
        ### neg_mean_absolute_error ini scoring parameter untuk nanti dapat di gunakan pada metrik mean_absolute_error
        ### Jika diatur ke 'raise', nilai error dimunculkan 

    print(model_cv_mae, i)

    score_mae.append(model_cv_mae)
    nilai_mean_mae.append(model_cv_mae.mean())
    nilai_std_mae.append(model_cv_mae.std())


    ## MAPE
    model_cv_mape = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_mean_absolute_percentage_error', 
        error_score='raise' 
        )
     ### neg_mean_absolute_percentage_error ini scoring parameter untuk nanti dapat di gunakan pada metrik mean_absolute_percentage_error
     ### Jika diatur ke 'raise', nilai error dimunculkan 

    print(model_cv_mape, i)

    score_mape.append(model_cv_mape)
    nilai_mean_mape.append(model_cv_mape.mean())
    nilai_std_mape.append(model_cv_mape.std())
```

```python
## Melakukan testing pada base model
models = {
    'Testing Linear Regression': LinearRegression(),
    'Testing KNN': KNeighborsRegressor(),
    'Testing Decision Tree Regressor': DecisionTreeRegressor(random_state=42),
    'Testing RandomForest': RandomForestRegressor(random_state=42),
    'Testing XGBoost': XGBRegressor(random_state=42)
}

score_mae = []
score_mape = []

## Melakukan prediksi pada test set
for i in models:

    model = Pipeline([
        ('preprocessing', transformer),
        ('model', models[i])
        ])

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    score_mae.append(mean_absolute_error(y_test, y_pred))
    score_mape.append(mean_absolute_percentage_error(y_test, y_pred))

score_before_tuning = pd.DataFrame({'MAE': score_mae, 'MAPE': score_mape}, index=models.keys())

## Evaluasi hasil dari 5 kandidat algoritma testing
print('Hasil Test Set (MAE dalam bentuk US Dollar, MAPE dalam bentuk persentase): ')
score_before_tuning
```

```python
## Kedalaman pohon
max_depth = list(np.arange(1, 11)) ## 1-10

## Learning rate
learning_rate = list(np.arange(1, 100)/100)

## Jumlah pohon
n_estimators = list(np.arange(100, 201))

## Jumlah baris tiap pohon (% dari total baris train set)
subsample = list(np.arange(2, 10)/10)

## Gamma (min_impurity_decrease)
gamma = list(np.arange(1, 11))

## Jumlah feature yang digunakan untuk tiap pohon (% dari total kolom train set)
colsample_bytree = list(np.arange(1, 10)/10)

## Alpha (regularization)
reg_alpha = list(np.logspace(-3, 1, 10))


## Hyperparam space XGboost
hyperparam_space_xgb = {
    'model__max_depth': max_depth, 
    'model__learning_rate': learning_rate,
    'model__n_estimators': n_estimators,
    'model__subsample': subsample,
    'model__gamma': gamma,
    'model__colsample_bytree': colsample_bytree,
    'model__reg_alpha': reg_alpha
}
```

```python
## Benchmark model dengan hyperparameter tuning yang dipilih : 
xgb = XGBRegressor(random_state=42, verbosity=0)
## verbosity = 0 artinya tidak menampilkan informasi progress hyperparameter tuning yang di running oleh program saat di train_test_split
## verbosity = 1 artinya  menampilkan informasi progress hyperparameter tuning (seperti epoch berapa, step berapa) yang di running oleh program saat di train_test_split
## verbosity = 2 artinya  menampilkan informasi progress hyperparameter tuning (1/64 Epoch) yang di running oleh program saat di train_test_split


## Membuat algorithm chains
estimator_xgb = Pipeline([
        ('preprocessing', transformer),
        ('model', xgb)
        ])

crossval = KFold(n_splits=5, shuffle=True, random_state=42) ## n_split 5 kombinasi, #shuffle untuk mengacak -> dibagi 5 berdasarkan n_splits

## Hyperparameter tuning
hyperparameter_tuning_xgb = RandomizedSearchCV(
    estimator_xgb,
    param_distributions = hyperparam_space_xgb,
    n_iter = 50, 
    cv = crossval, 
    scoring = ['neg_mean_absolute_error', 'neg_mean_absolute_percentage_error'], 
    n_jobs = -1,
    refit = 'neg_mean_absolute_percentage_error',
    random_state = 42
)
```
